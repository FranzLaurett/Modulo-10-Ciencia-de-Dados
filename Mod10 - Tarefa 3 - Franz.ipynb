````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python

df_raw = pd.read_csv('previsao_de_renda.csv')

df_raw.head()
````

````python
df = df_raw.copy()

df.drop(['Unnamed: 0', 'mau', 'index'], axis = 1, inplace = True)

df.head()
````

````python
#1 - construindo gráfico da variável qualitativa 'sexo' / tempo

sns.set(rc={'figure.figsize':(12,7)})
ax = sns.countplot(x='data_ref',  hue = 'sexo', data=df_raw)

tick_labs = df_raw['data_ref'].unique()

ticks = ax.set_xticks(list(range(df_raw['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
title = ax.set_title('Sexo')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

Ao avaliar a estabilidade na distribuição da variável 'sexo' ao longo do tempo, é possível observar que pouco se alterou no período. A distribuição da população do sexo feminino sempre se manteve acima da masculina.
